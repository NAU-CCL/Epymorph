```python
import numpy as np
import matplotlib.pyplot as plt

from epymorph.util import long_range_von_mises_distribution


tau_values = np.linspace(0, 2 * np.pi, 1000)

sij = 0.8
phiij = np.pi
deltaij = 0.3

probability_values = long_range_von_mises_distribution(sij, phiij, deltaij, tau_values)
probability_values /= np.max(probability_values)

plt.plot(tau_values, probability_values)
plt.title('von Mises Distribution')
plt.xlabel('T')
plt.ylabel('t(ij)')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', '$\\frac{\pi}{2}$', '$\pi$', '$\\frac{3\pi}{2}$', '$2\pi$'])
plt.show()
```
