### SIRH 

An advance model of the SIR model as we add a new compartment which is hospitalized. Individuals who are infected can either be hospitalized or remain infected. 


### TOML file params 

theta = 0.1 

beta = 0.5739426782

move_control = 0.9

infection_duration = 4.0

immunity_duration = 90.0

infection_seed_loc = 0

infection_seed_size = 10000

hospitalization_rate = 0.25

hospitalization_duration = 4.0



# SIRH with no movement

### plots SIRH model by prevalence
python -m epymorph run --ipm sirh --mm no --geo pei --params "./scratch/sirh_params.toml" --start_date "2015-01-01" --duration "150d" --chart "p0"
### creats SIRH.csv file
python -m epymorph run --ipm sirh --geo pei --mm no --start_date 2015-01-01 --duration 150d --params ./scratch/sdh_params.toml --out ./scratch/sirh_no.csv
