Running the SPARSEMOD IPM.

```python
from datetime import date

from epymorph.data import geo_library as geo
from epymorph.data import ipm_library as ipm
from epymorph.data import mm_library as mm
from epymorph.run import plot_pop
from epymorph.simulation import Simulation

sim = Simulation(geo['single_pop_geo'](), ipm['sparsemod'](), mm['no']())

out = sim.run(
    param={
        'infection_seed_loc': 0,
        'infection_seed_size': 10000,
        'beta_1': 0.3,
        'omega_1': 0.55,
        'omega_2': 0.05,
        'delta_1': 0.333,
        'delta_2': 0.5,
        'delta_3': 0.166,
        'delta_4': 0.142,
        'delta_5': 0.125,
        'gamma_a': 0.166,
        'gamma_b': 0.333,
        'gamma_c': 0.25,
        'rho_1': 0.4,
        'rho_2': 0.175,
        'rho_3': 0.015,
        'rho_4': 0.2,
        'rho_5': 0.6,
    },
    start_date=date(2015, 1, 1),
    duration_days=150
)

plot_pop(out, 0)
```
