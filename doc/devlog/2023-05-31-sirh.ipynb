## SIRH 

An advance model of the SIR model as we add a new compartment which is hospitalized. Individuals who are infected can either be hospitalized or remain infected. 


## TOML file params 

theta = 0.1 

beta = 0.5739426782

move_control = 0.9

infection_duration = 4.0

immunity_duration = 90.0

infection_seed_loc = 0

infection_seed_size = 10000

hospitalization_rate = 0.25

hospitalization_duration = 4.0



### SIRH with no movement

python -m epymorph run --ipm sirh --mm no --geo pei --params "./scratch/params.toml" --start_date "2015-01-01" --duration "150d" --chart "e0"
### Incidence

```python
import os

from epymorph.run import run

dir = os.path.expanduser('~/Desktop/Github/Epymorph/scratch')

exit_code = run(
    ipm_name = 'sirh',
    mm_name = 'no',
    geo_name = 'pei',
    start_date_str = '2015-01-01',
    duration_str = '150d',
    params_path = f"{dir}/params.toml",
    out_path =None,
    chart = 'e0',
    profiling=False
)
```
