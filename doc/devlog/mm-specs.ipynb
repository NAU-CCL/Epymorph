Let's try running our different movement models with different models. 

### Weekend movement

Movement wiil only happen on weekends. The duration of their move will be the length of the weekend. 
python -m epymorph run --ipm pei --mm weekend --geo pei --params "./scratch/sirh_params.toml" --start_date "2015-01-01" --duration "150d" --chart "e0"

```python
import os

from epymorph.run import run

dir = os.path.expanduser('~/Desktop/Github/Epymorph/scratch')

exit_code = run(
    ipm_name = 'pei',
    mm_name = 'weekend',
    geo_name = 'pei',
    start_date_str = '2015-01-01',
    duration_str = '150d',
    params_path = f"{dir}/sirh_params.toml",
    out_path = None,
    chart = 'e0',
    profiling=False
    
)
```
